```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt

# Path to the test image
test_image_path = r'C:\Users\RISHI PANDE\Documents\Prediction\train_Nails\Splinter Hemmorrage\Splinter Hemmorrage (7).PNG'

# Path to the trained model
model_path = r'C:\Users\RISHI PANDE\Documents\Prediction\nail_disease_model.h5'  # Update this with the actual path to your model

# Load the trained model
model = load_model(model_path)

# Image dimensions
img_size = (224, 224)

# Load and preprocess the test image
img = load_img(test_image_path, target_size=img_size)
img_array = img_to_array(img) / 255.0  # Normalize the image to [0, 1]
img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension

# Predict the class of the test image
predictions = model.predict(img_array)
predicted_class = np.argmax(predictions, axis=1)[0]

# Class labels (assuming these were used during training)
class_labels = ['Bluish nail', 'Healthy', 'Splinter Hemorrhage', 'White Nail', 'Yellow Nails']

# Display the prediction
predicted_label = class_labels[predicted_class]
print(f"Predicted class: {predicted_label}")
print(f"Predicted probabilities: {predictions[0]}")

# Output specific messages based on the predicted class
if predicted_label == 'Bluish nail':
    message = "The following person has less oxygen."
elif predicted_label == 'Healthy':
    message = "The following person is healthy."
elif predicted_label == 'Splinter Hemorrhage':
    message = "The following person is suffering from trauma."
elif predicted_label == 'White Nail':
    message = "The following person is suffering from kidney disease."
elif predicted_label == 'Yellow Nails':
    message = "The following person is suffering from jaundice."

print(message)

# Display the test image
plt.imshow(img)
plt.title(f"Predicted: {predicted_label}")
plt.axis('off')
plt.show()
```
